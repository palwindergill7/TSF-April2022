# Import Libraries


```python
import pandas as pp
import numpy as nn
import matplotlib.pyplot as ppt 
import seaborn as ss
```

```python
from sklearn import datasets
from sklearn.cluster import KMeans
```

**Importing Dataset**

```python
iris = datasets.load_iris()
iris_df = pp.DataFrame(iris.data, columns = iris.feature_names)
```

```python
x = iris_df.iloc[:, [0, 1, 2, 3]].values
```

```python
wcss=[]
for i in range(1, 11):  
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
    kmeans.fit(x)  
    wcss.append(kmeans.inertia_)  
```

```python
ppt.plot(range(1, 11), wcss,'o-')
ppt.title('The elbow method')
ppt.xlabel('Number of clusters')
ppt.ylabel('WCSS') 
ppt.show()
```

**Creating the kmeans classifier**

```python
kmeans = KMeans(n_clusters = 3, init = 'k-means++',
 max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

**Visualising the clusters**

```python
ppt.figure(figsize=(15,7.5))
ppt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1],
 s = 100, c = 'red', label = 'Iris-setosa')
ppt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1],
 s = 100, c = 'blue', label = 'Iris-versicolour')
ppt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1],
 s = 100, c = 'green', label = 'Iris-virginica')
ppt.scatter(kmeans.cluster_centers_[:,0],
kmeans.cluster_centers_[:,1],
s = 300, c = 'black', label = 'Centeroids')

ppt.legend()
ppt.show()
```
